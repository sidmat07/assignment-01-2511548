```python
from sentence_transformers import SentenceTransformer
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sentences = [
    # Cricket
    "The batsman scored a century in the match",
    "The bowler took three wickets in one over",
    "India won the match",

    # Cooking
    "I cooked a delicious meal",
    "My wife baked the cake",
    "Spices are necessity for cooking Indian food",

    # Cybersecurity
    "Strong passwords help protect your accounts",
    "Encryption ensures data security",
    "Firewalls prevent unauthorized access",
    "Phishing attacks can steal sensitive information"
]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
```

```python
similarity_matrix = cosine_similarity(embeddings)

plt.figure(figsize=(10,8))
sns.heatmap(similarity_matrix, annot=False, cmap='coolwarm')
plt.title("Cosine Similarity Matrix")
plt.show()
```

```python
query = "The bowler took three wickets in one over"
query_embedding = model.encode([query])

similarities = cosine_similarity(query_embedding, embeddings)[0]

# Get top 2 similar sentences
top_indices = similarities.argsort()[-2:][::-1]

for idx in top_indices:
  print(f"Sentence: {sentences[idx]}")
  print(f"Similarity Score: {similarities[idx]:.4f}")
  print()
```
